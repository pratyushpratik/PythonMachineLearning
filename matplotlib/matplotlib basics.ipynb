## video 3 (plotting ads data with stackplot)

```python
myDays = [1,2,3,4,5]
```

```python
organic = [8, 9, 8, 3, 3]
ads = [8,9,7,12,8]
affiliate = [3,5,5,4,3]
seller = [9,5,8,9,14]
```

```python
plt.stackplot(myDays, organic,ads,affiliate,seller)
```

```python
plt.stackplot(myDays, organic,ads,affiliate,seller, colors=['#F97F51', '#1b9cfc', '#f8efba', '#ff3366'])
```

```python
plt.plot([], [], color='#F97F51', label ='organic')
plt.plot([], [], color='#1b9cfc', label ='ads')
plt.plot([], [], color='#f8efba', label ='affiliate')
plt.plot([], [], color='#ff3366', label ='seller')
plt.stackplot(myDays, organic,ads,affiliate,seller, colors=['#F97F51', '#1b9cfc', '#f8efba', '#ff3366'])
plt.xlabel('Number of days')
plt.ylabel('Sales')
plt.legend()
```
